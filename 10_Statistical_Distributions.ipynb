Key Concepts:

- A sample of data will form a distribution, and by far the most well-known distribution is the Gaussian distribution, often called the Normal distribution.
- This distribution describes the grouping or the density of the observations, called the **probability density function**.
- A summary of these relationships between observations is called a **cumulative density function**.

## Distributions

From a practical perspective, we can think of a distribution as a function that describes the relationship between observations in a sample space.

**Example**: Study of human ages, with ages representing observations in the domain, like age 0 to 125. So the distribution is a mathematical function that describes the relationship of observations of different heights.

Many data conform to well-known and well-understood mathematical functions, such as the Gaussian distribution. A function can *fit* the data with parameterized *mean* and *standard deviation* (in Gaussian). So we need to come up with a distribution function to calculate the likelihood of observations, plotting the relationship.

### Density Functions

Distributions are often described in terms of their density or density functions. Density functions are functions that describe how the proportion of data or likelihood of the proportion of observations change over the range of the distribution. Two types of density functions:

- **Probability Density function (PDF)**: Calculate the probability of observing a given value.
- **Cumulative Density function (CDF)**: Calculate the probability of an observation equal or less than a value.

Both PDF and CDF arae *continuous* functions. The equivalent of a PDF for a discrete distribution is called a probability mass function (PMF).


## Gaussian Distribution

A Gaussian distribution can be described using two parameters:

- **mean**: Denoted with the Greek lowercase letter mu ($\mu$), is the expected value of the distribution.
- **variable**: Denoted with the Greek lowercase letter sigma ($\sigma^2$) raised to the second power (because the units of the variable are squared), describes the spread of observation from
the mean.

It is common to use normalized calculation of the variance, like

- **standard deviation**: Denoted with the Greek lowercase letter sigma ($\sigma$), describes the normalized spread of observations from the mean.

We can work with the Gaussian distribution via the **norm SciPy** module. The **norm.pdf()** function can be used to create a Gaussian probability density function with a given sample space, mean, and standard deviation.

**mean(x):**

$\mu = \frac{1}{n} \times \displaystyle\sum_{i=1}^{n}x_i$

**variance(x):**

$\sigma^2 = \frac{1}{n-1} \times \displaystyle\sum_{i=1}^{n}(x_i - \mu)^2$

**standard deviation(x):**

$\sigma = \sqrt{\sigma^2}$

or

$\sigma = \sqrt{\frac{1}{n-1} \times \displaystyle\sum_{i=1}^{n}(x_i - \mu)^2}$

```python
# example below creates a Gaussian PDF with a sample space from -5 to 5, 
# mean: 0
# standard deviation: 1

# plot the Gaussian PDF
from numpy import arange
from matplotlib import pyplot
from scipy.stats import norm

# define the distribution parameters
sample_space = arange(-5, 5, 0.001)
mean = 0
std_dev = 1.0

# calculate the PDF
pdf = norm.pdf(sample_space, mean, std_dev)

# plot
pyplot.xlabel('Sample space');
pyplot.ylabel('Likelihood of each value')
pyplot.plot(sample_space, pdf)
#pyplot.show(0)
```

```python
# The norm.cdf() function can be used to create a Gaussian cumulative density function.
# S-shape

# plot the Gaussian CDF
from numpy import arange
from matplotlib import pyplot
from scipy.stats import norm

# define the distribution parameters
sample_space = arange(-5, 5, 0.001)

# calculate the CDF
cdf = norm.cdf(sample_space)

# plot 
pyplot.xlabel('Sample space');
pyplot.ylabel('Cumulative probability')
pyplot.plot(sample_space, cdf)
pyplot.show()
```

## Student's t-Distribution

It is a distribution that arises when attempting to estimate the mean of a normal distribution with different sized samples. As such, it is a helpful shortcut when describing uncertainty or error related to estimating population statistics for data drawn from Gaussian distributions when the size of the sample must be taken into account.

The distribution can be described using a single parameter:

- **number of degrees of freedom**: denoted with the lowercase Greek letter nu ($\nu$), denotes the number degrees of freedom. It describes the number of pieces of information used to describe a population quantity.
- **Reference**: https://blog.minitab.com/en/statistics-and-quality-data-analysis/what-are-degrees-of-freedom-in-statistics#:~:text=Degrees%20of%20freedom%20are%20often,vary%20when%20estimating%20statistical%20parameters.

Where x is the observations from the Gaussian distribution, mean is the average observation of $x$, S is the standard deviation and n is the total number of observations. The resulting observations form the t-observation with (n − 1) degrees of freedom. In practice, if you require a value from a t-distribution in the calculation of a statistic, then the number of degrees of freedom will likely be n − 1, where n is the size of your sample drawn from a Gaussian distribution.

$\nu = \displaystyle\frac{x - mean(x)}{\displaystyle\frac{S}{\sqrt{n}}}$

or

$\nu = \displaystyle\frac{x - \mu}{\frac{\sigma}{\sqrt{n}}}$

```python
# SciPy provides tools for working with the t-distribution in the stats.t module. 
# The t.pdf() function can be used to create a Student’s t-distribution with the specified degrees of freedom.

# plot the t-distribution pdf
from numpy import arange
from matplotlib import pyplot
from scipy.stats import t

# define the distribution parameters
sample_space = arange(-5, 5, 0.001)
degree_of_freedom = len(sample_space) - 1
print(degree_of_freedom)

# calculate the PDF
pdf = t.pdf(sample_space, degree_of_freedom)

# plot
pyplot.xlabel('Sample space');
pyplot.ylabel('Likelihood of each value')
pyplot.plot(sample_space, pdf)
pyplot.show()
```

Running the example creates and plots the t-distribution PDF. We can see the familiar bell-shape to the distribution much like the normal. A key difference is the fatter tails in the distribution (hard to see by eye), highlighting the increased likelihood of observations in the tails compared to that of the Gaussian.

```python
# The t.cdf() function can be used to create the cumulative density function for the t- distribution.

# plot the t-distribution cdf
from numpy import arange
from matplotlib import pyplot
from scipy.stats import t

# define the distribution parameters
sample_space = arange(-5, 5, 0.001)
degree_of_freedom = len(sample_space) - 1

# calculate the CDF
cdf = t.cdf(sample_space, degree_of_freedom)

# plot
pyplot.xlabel('Sample space');
pyplot.ylabel('Cumulative probability')
pyplot.plot(sample_space, cdf)
pyplot.show()
```

## Chi-Squared Distribution

The Chi-Squared distribution is denoted as the lowercase Greek letter chi ($\chi$) pronounced “ki” as in “kite”, raised to the second power ($\chi^2$). It is easier to write Chi-Squared. The Chi-Squared distribution has one parameter:

- degrees of freedom, denoted $k$.

$chi = \displaystyle\sum_{i=1}^{k}x_i^2$

or 

$\chi^2 = \displaystyle\sum_{i=1}^{k}x_i^2$



```python
# SciPy provides the stats.chi2 module for calculating statistics for the Chi-Squared distri- bution.

# plot the chi-squared pdf
from numpy import arange
from matplotlib import pyplot
from scipy.stats import chi2

# define the distribution parameters
sample_space = arange(0, 50, 0.01)
degree_of_freedom = 20

# calculate the pdf
pdf = chi2.pdf(sample_space, degree_of_freedom)

# plot
pyplot.xlabel('Sample space');
pyplot.ylabel('Likelihood of each value with 20 degree of freedom')
pyplot.plot(sample_space, pdf)
pyplot.show()

# With 20 degrees of freedom, we can see that the expected value of the distribution is just short of the value 20 on the sample space.
```

```python
# The chi2.cdf() function can be used to calculate the cumulative density function over the same sample space.

from numpy import arange
from matplotlib import pyplot
from scipy.stats import chi2

# define the distribution parameters
sample_space = arange(0, 50, 0.01)
degree_of_freedom = 20

# calculate the CDF
cdf = chi2.cdf(sample_space, degree_of_freedom)

# plot
pyplot.xlabel('Sample space');
pyplot.ylabel('Cumulative probability')
pyplot.plot(sample_space, cdf)
pyplot.show()

# The distribution helps to see the likelihood for the Chi-Squared value around 20 with the fat tail to the right of the distribution that would continue on long after the end of the plot.
```
